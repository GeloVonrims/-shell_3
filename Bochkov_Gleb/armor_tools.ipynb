### Раскадровка

```python
import cv2
import os

# Указываем имя видеофайла
video_path = 'data_v_train.mp4'

# Открываем видеофайл
cap = cv2.VideoCapture(video_path)

# Получаем частоту кадров видео
frame_rate = int(cap.get(cv2.CAP_PROP_FPS))

# Устанавливаем интервал между кадрами в 1 секунду
frame_interval = 1 * frame_rate

# Устанавливаем интервал между кадрами в 3 секунды
# frame_interval = 3 * frame_rate

# Присваеваем имя папки для сохранения раскадрованных изображений
output_folder = 'frames'

# Очищаем выходную папку если она уже существует
!rm -r /content/{output_folder}

# Создаем папку для сохранения раскадрованных изображений если ее еще нет
os.makedirs(output_folder, exist_ok=True)

# Переменная для отслеживания номера кадра
frame_count = 0
```

```python
# Копируем с Google Disk видеофайл для доступа в Colab
!cp -n /content/drive/MyDrive/data/data_v_train.mp4 /content
```

```python
# Читаеме кадры видео и сохраняем их с заданным интервалом
while True:
    # Читаем следующий кадр
    ret, frame = cap.read()

    # Если чтение кадра не удалось, выходим из цикла
    if not ret:
        break

    # Проверяем, является ли текущий кадр кадром, который нужно сохранить
    if frame_count % frame_interval == 0:
        # Формируем путь к файлу для сохранения кадра
        frame_path = os.path.join(output_folder, f'frame_{frame_count:06d}.png')

        # Сохраняем кадр в формате PNG
        cv2.imwrite(frame_path, frame)

    # Увеличиваем номер кадра
    frame_count += 1

# Закрываем видеофайл
cap.release()

# Выводим сообщение о завершении раскадровки видео
print('Раскадровка видео завершена.')
```

### Вырезать кусок

```python
import cv2

# Укажите путь к исходному видеофайлу
input_video_path = 'data_v_train.mp4'

# Укажите путь для сохранения вырезанного куска
output_video_path = 'output_video.mp4'

# Установите временные метки начала и конца куска (в секундах)
start_time = 10  # Начало куска
end_time = 20    # Конец куска

# Откройте исходное видеофайл
cap = cv2.VideoCapture(input_video_path)

# Получите информацию о видео (размер кадров, FPS и т.д.)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = cap.get(cv2.CAP_PROP_FPS)
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

# Вычислите временные метки начала и конца в кадрах
start_frame = int(start_time * frame_rate)
end_frame = int(end_time * frame_rate)

# Проверьте, что указанные временные метки находятся в пределах видео
if start_frame >= total_frames or end_frame >= total_frames:
    raise ValueError('Временные метки превышают общее количество кадров в видео.')

# Установите позицию видеофайла на начальную временную метку
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)

# Создайте объект VideoWriter для сохранения вырезанного куска
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_video = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (frame_width, frame_height))

# Читайте кадры и сохраняйте их в выходной видеофайл до достижения конечной временной метки
frame_count = start_frame
while frame_count <= end_frame:
    # Читайте кадр из исходного видео
    ret, frame = cap.read()

    # Если чтение кадра не удалось, выходим из цикла
    if not ret:
        break

    # Сохраните кадр в выходной видеофайл
    output_video.write(frame)

    # Увеличивайте номер кадра
    frame_count += 1

# Закройте видеофайлы
cap.release()
output_video.release()

# Выведите сообщение о завершении вырезания куска из видео
print('Вырезание куска из видео завершено.')
```

### Склеить куски

```python
import cv2

# Укажите пути к исходным видеофайлам
video1_path = 'video1.mp4'
video2_path = 'video2.mp4'

# Укажите путь для сохранения объединенного видеофайла
output_video_path = 'output_video.mp4'

# Откройте первый видеофайл
cap1 = cv2.VideoCapture(video1_path)

# Получите информацию о первом видео (размер кадров, FPS и т.д.)
frame_width = int(cap1.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap1.get(cv2.CAP_PROP_FRAME_HEIGHT))
frame_rate = cap1.get(cv2.CAP_PROP_FPS)

# Создайте объект VideoWriter для сохранения объединенного видеофайла
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
output_video = cv2.VideoWriter(output_video_path, fourcc, frame_rate, (frame_width, frame_height))

# Читайте и записывайте кадры из первого видеофайла
while True:
    ret, frame = cap1.read()
    if not ret:
        break
    output_video.write(frame)

# Закройте первый видеофайл
cap1.release()

# Откройте второй видеофайл
cap2 = cv2.VideoCapture(video2_path)

# Читайте и записывайте кадры из второго видеофайла
while True:
    ret, frame = cap2.read()
    if not ret:
        break
    output_video.write(frame)

# Закройте второй видеофайл
cap2.release()

# Закройте объект VideoWriter
output_video.release()

# Выведите сообщение о завершении склеивания видео
print('Склеивание видео завершено.')
```
